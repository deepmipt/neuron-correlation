### Method `TestTrainRepeatedly.test_dataset_tensor_saving()`

```python
import os

import numpy as np
import tensorflow as tf
import tests.integration.test_train_repeatedly as module


ZEROS_ONES_CREATION = {
    "tensor1": {
        "module": "tensorflow",
        "function": "zeros",
        "args": [[2, 9]]
    },
    "tensor2": {
        "module": "tensorflow",
        "function": "ones",
        "args": [[3, 5]]
    }
}


t1 = tf.zeros([2, 9])
t2 = tf.ones([3, 5])

s1 = tf.summary.tensor_summary('tensor1', t1)
s2 = tf.summary.tensor_summary('tensor2', t2)

path = os.path.expanduser(
    '~/neuron-correlation/results_of_tests/integration/test_mnist/train_repeatedly/train_tensor_summary')

with tf.Session() as sess:
    for j in range(10):
        logdir1 = os.path.join(path, str(j), 'train_tensors/tensor1')
        logdir2 = os.path.join(path, str(j), 'train_tensors/tensor2')
        w1 = tf.summary.FileWriter(logdir=logdir1)
        w2 = tf.summary.FileWriter(logdir=logdir2)
        for i in range(0, 150001, 10000):
            s_1 = sess.run(s1)
            w1.add_summary(s_1, global_step=i)
        w1.flush()
        step = 0
        factor = 1.3
        while step < 150000:
            s_2 = sess.run(s2)
            w2.add_summary(s_2, global_step=step)
            new_step = int(np.ceil(step*factor))
            if new_step == step:
                new_step += 1
            step = new_step
        s_2 = sess.run(s2)
        w2.add_summary(s_2, global_step=150000)
        w2.flush()
```

```python
import os

import numpy as np
import tensorflow as tf
import tests.integration.test_train_repeatedly as module


def custom_logrange(start, stop, factor, include_stop=False):
    values = []
    while start < stop:
        values.append(start)
        if start >= int(start * factor):
            start += 1
        else:
            start = int(start * factor) + int(start * factor % 1 > 0)
    if include_stop:
        values.append(stop)
    return values


path = os.path.expanduser(
    '~/neuron-correlation/results_of_tests/integration'
    '/test_mnist/train_repeatedly/dataset_tensors/0/dataset_tensors'
)

tensor_values = module.get_summary_tensors_values(ZEROS_ONES_CREATION)
tensor_steps = {
    "tensor1": {
        "train": list(range(0, 150000, 5000)),
        "datasets": list(range(9999))
    },
    "tensor2": {
        "train": custom_logrange(0, 150000, 1.3, True),
        "dataset": list(range(0, 100, 3))
    }
}

report = module.check_ds_summarized_in_launch_dir(
    DS_ZEROS_ONES,
    path,
    tensor_values,
    tensor_steps,
)
```
